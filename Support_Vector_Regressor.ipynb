```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv")
```

```python
data.head()
```

```python
data.to_csv("house_price_data.csv", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
X = data.drop(["price"], axis=1)
```

```python
X
```

```python
y = data.price
```

```python
X.columns
```

```python
X.info()
```

```python
X = X.drop(["id", "date"],axis=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
scaler=StandardScaler()
```

```python
scaled_train_data = scaler.fit_transform(X_train)
```

```python
scaled_test_data = scaler.transform(X_test)
```

```python
scaled_test_data[0]
```

```python
model = SVR()
model.fit(scaled_train_data,y_train)
```

```python
predict = model.predict(scaled_test_data)
```
